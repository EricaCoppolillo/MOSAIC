# Moral Dilemmas

```python
raw_data_folder = os.path.join(data_folder, "raw_data")
```

```python
def process_raw_data(raw_data, test=None):

    list_scenarios = []

    for x in raw_data:
        for scenario in x:
            list_scenarios.append(tuple(scenario))

    scenarios_without_duplicates = list(set(list_scenarios))

    print(f"{len(scenarios_without_duplicates)} distinct scenarios")

    questions = []

    if test != "tinker_tots":
    
        for scenario in scenarios_without_duplicates:
            scenario_A, scenario_B = scenario
    
            question = f'''Scenario A:\n\n{scenario_A}\n\n\nScenario B:\n\n{scenario_B}'''
            questions.append(question)
    else:
        letters = ["A", "B", "C", "D", "E", "F", "G", "H"]
        
        for scenario in scenarios_without_duplicates:
            question = ""
            for i in range(len(scenario)):
                question += f'''Scenario {letters[i]}:\n\n{scenario[i]}'''
                if i < len(scenario) - 1:
                    question += "\n\n\n"
            
            questions.append(question)

    return questions
```
